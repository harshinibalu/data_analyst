```python
#concat,merge and join 

data1 = {
    'ID':['1','2','3','4','5'],
    'Feature1': ['A','B','C','D','E'],
    'Feature2': ['L','M','N','O','P']
}

df1 = pd.DataFrame(data1, columns = ['ID','Feature1','Feature2'])


data2 = {
    'ID':['1','3','6','7','8'],
    'Feature1': ['AB','BC','CD','DE','EF'],
    'Feature2': ['KL','LM','MN','NO','OP']
}

df2 = pd.DataFrame(data2, columns = ['ID','Feature1','Feature2'])
```

```python
print(df1)
print(df2)
```

```python
#axis =0 means rows ,axis =1 means column   

#print(pd.concat([df1,df2],axis=0))
print(pd.concat([df1,df2])) # this also implement axis =0
```

```python
#column wise
print(pd.concat([df1,df2],axis=1))
```

```python
#concat,merge and join 
#column name changed so the result changes

data1 = {
    'ID':['1','2','3','4','5'],
    'Feature1': ['A','B','C','D','E'],
    'Feature2': ['L','M','N','O','P']
}

df1 = pd.DataFrame(data1, columns = ['ID','Feature1','Feature2'])


data2 = {
    'ID':['1','3','6','7','8'],
    'Feature1': ['AB','BC','CD','DE','EF'],
    'Feature3': ['KL','LM','MN','NO','OP']
}

df2 = pd.DataFrame(data2, columns = ['ID','Feature1','Feature3'])
```

```python
print(pd.concat([df1,df2])) #nan - not a number
```

```python
#merge -->on - same datatype 

a = pd.merge(df1,df2, on='ID',how='inner')
print(a)
```

```python
#outer merge
a = pd.merge(df1,df2, on='ID',how='outer')
print(a)
```

```python
#left merge
a = pd.merge(df1,df2, on='ID',how='left')
print(a)
```

```python
#right merge
a = pd.merge(df1,df2, on='ID',how='right')
print(a)
```

```python
#for removng duplicates
b= pd.merge(df1,df2, on='ID',how='outer',indicator=True)
print(b)
```

```python
b[b['_merge']!='both']
```

```python
#we can use the above for our best this we dont use for exact matching only it will work correctly
c=pd.concat([df1,df2],axis=1).drop_duplicates(keep=False)
print(c)
```

```python
#join - lsuffix - left suffix if column name is  matching on both side means we need to put like this

df1.join(df2, lsuffix="f1", rsuffix="f2")
```

```python
#only added id from f2 as follows
df1.join(df2["ID"], lsuffix="f1", rsuffix="f2")
```

```python
df1.join(df2["ID"], lsuffix="f1", rsuffix="f2").join(df2["Feature1"], lsuffix="f1", rsuffix="f2")
```
