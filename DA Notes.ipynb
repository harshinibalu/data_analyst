```python
#without encapsulation
class bankaccount():
    def __init__(self,actnum,bal):
        self.actnum = actnum
        self.bal = bal #public attribute

    def deposit(self,amt):
      self.bal = self.bal+amt
      return self.bal
    
    def withdraw(self,amt):
       if amt > self.bal:
          print("Insufficient balance")
          return
       self.bal = self.bal - amt
       return self.bal
    def get_bal(self):
       return self.bal
account = bankaccount("123456789",100)
print(account.get_bal())  #100

#with the bal being public,we can change it directly ,potentially causing data inconsistency or misuse
account.bal = 1000000
print(account.get_bal()) #1000000

#Or even set it to nonsensical values:
account.bal = -500
print(account.get_bal()) #-500
    
```

```python
#with encapsulation/abstraction -- see __ before bal is encapsulation hiding we cannot chamge it by directly adding
#only by deposit and withdrawal only we can change the balance..not by simply adding
class bankaccount():
    def __init__(self,actnum,bal):
        self.actnum = actnum
        self.__bal = bal #private attribute

    def deposit(self,amt):
      self.__bal += amt
      return self.__bal
    
    def withdraw(self,amt):
       if amt > self.__bal:
          print("Insufficient balance")
          return
       self.__bal -=  amt
       return self.__bal
    

    def get_bal(self):
       return self.__bal
    
account = bankaccount("123456789",100)
print(account.get_bal())  #100

#with the bal being public,we can change it directly ,potentially causing data inconsistency or misuse
account.bal = 1000000
print(account.get_bal()) #1000000

#Or even set it to nonsensical values:
account.bal = -500
print(account.get_bal()) #-500

account.deposit(500)
print(account.get_bal()) 
    
```

SQL is a query language.
MySQL is a software that implements SQL to manage databases



SQL is a language, while MySQL is a database management system.
SQL is used by a variety of RDBMSs (like MySQL, PostgreSQL, SQL Server, Oracle), whereas MySQL is a specific software that uses SQL as its query language.
SQL focuses on querying and managing data, while MySQL is the system that stores and organizes that data.


MySQL:
What it is: MySQL is an RDBMS (Relational Database Management System) that uses SQL as its query language. It is an open-source, widely-used software that manages databases and provides the environment to store and manage relational data.

Functions:

MySQL is software used to create, manage, and interact with databases, and it uses SQL to query and manipulate the data stored in those databases.
It handles things like data integrity, transactions, security, backup, and recovery.
Usage:

You install MySQL on a server, create databases, and then use SQL to interact with those databases.
MySQL allows you to run SQL queries, but it also comes with additional tools and features for managing databases, such as:
MySQL Workbench for database design and administration.
MySQL Cluster for high-availability applications.
